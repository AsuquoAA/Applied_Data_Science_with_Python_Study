### Question 2
What is the class distribution? (i.e. how many instances of `malignant` and how many `benign`?)

*This function should return a Series named `target` of length 2 with integer values and index =* `['malignant', 'benign']`

```python
def answer_two():
    dataframe = answer_one()
    counts = dataframe['target'].value_counts()
    counts = counts.reindex([0,1])
    counts.index = cancer['target_names']
    return counts
    # YOUR CODE HERE
    raise NotImplementedError()
answer_two()    
```

### Question 3
Split the DataFrame into `X` (the data) and `y` (the labels).

*This function should return a tuple of length 2:* `(X, y)`*, where* 
* `X` *has shape* `(569, 30)`
* `y` *has shape* `(569,)`.

```python
def answer_three():
    dataframe = answer_one()
    X = dataframe.drop('target',axis=1)
    y = dataframe['target']
    return X, y
    # raise NotImplementedError() 
answer_three()    
```

### Question 4
Using `train_test_split`, split `X` and `y` into training and test sets `(X_train, X_test, y_train, and y_test)`.

**Set the random number generator state to 0 using `random_state=0` to make sure your results match the autograder!**

*This function should return a tuple of length 4:* `(X_train, X_test, y_train, y_test)`*, where*

```python
from sklearn.model_selection import train_test_split

def answer_four():
    X,y = answer_three()
    X_train, X_test, y_train, y_test = train_test_split(X,y,random_state=0)
    return X_train, X_test, y_train, y_test
    # YOUR CODE HERE
    raise NotImplementedError()
answer_four()    
```

### Question 5
Using KNeighborsClassifier, fit a k-nearest neighbors (knn) classifier with `X_train`, `y_train` and using one nearest neighbor (`n_neighbors = 1`).

*This function should return a `sklearn.neighbors.classification.KNeighborsClassifier`.

```python
from sklearn.neighbors import KNeighborsClassifier

def answer_five():
    X_train, X_test, y_train, y_test = answer_four()
    knn = KNeighborsClassifier(n_neighbors=1)
    knn.fit(X_train,y_train)
    return knn
    # YOUR CODE HERE
    raise NotImplementedError()
answer_five()    
```

### Question 8

### Optional plot

Try using the plotting function below to visualize the different predicition scores between train and test sets, as well as malignant and benign cells.

```python
def accuracy_plot():
    import matplotlib.pyplot as plt
    
    %matplotlib inline
    X_train, X_test, y_train, y_test = answer_four()
    knn = answer_five()
    train_accuracy = knn.score(X_train,y_train)
    test_accuracy = knn.score(X_test,y_test)
    train_malignant = np.mean(knn.predict(X_train[y_train==0])==y_train[y_train==0])
    train_benign = np.mean(knn.predict(X_train[y_train==1])==y_train[y_train==1])
    test_malignant = np.mean(knn.predict(X_test[y_test==0])==y_test[y_test==0])
    test_benign = np.mean(knn.predict(X_test[y_test==1])==y_test[y_test==1])
    groups = ['Train Malignant', 'Train Benign', 'Test Malignant', 'Test Benign']
    accuracies = [train_malignant, train_benign, test_malignant, test_benign]
    
    plt.figure(figsize=(8,5))
    plt.bar(groups,accuracies,color=['blue','red','purple','yellow'])
    plt.ylim(0,1)
    plt.xlabel('Group')
    plt.ylabel('Accuracy')
    plt.title('Prediction Accuracy for Train/Test and Malignant/Benign Cells')
    plt.show()
    # YOUR CODE HERE
    #raise NotImplementedError()
accuracy_plot()    
```
