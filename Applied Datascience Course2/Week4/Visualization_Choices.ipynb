## __Matplotlib vs Seaborn vs Pandas.plot()__

| Library              | Best for                                      | Pros                                      | Cons                                      |
|---------------------|-----------------------------------------------|-------------------------------------------|-------------------------------------------|
| Matplotlib           | Customizable, complex visualizations          | Flexible, powerful, highly detailed      | Requires more code to achieve good visuals |
| Seaborn             | Statistical visualizations, data exploration   | Attractive, high-level, quick setup      | Limited customization, dependent on Matplotlib |
| Pandas.plot()      | Quick, simple plots from DataFrames or Series | Fast, straightforward, minimal setup     | Limited flexibility, basic visualizations  |


```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# Sample data
data = {
    'Date': pd.date_range(start='2020-01-01', periods=365, freq='D'),
    'Temperature': np.random.normal(loc=20, scale=5, size=365),
    'Precipitation': np.random.gamma(shape=2, scale=1, size=365)
}

df = pd.DataFrame(data)

# Create a figure with two subplots
fig, ax1 = plt.subplots()

# Temperature plot
ax1.set_xlabel('Date')
ax1.set_ylabel('Temperature (°C)', color='tab:red')
ax1.plot(df['Date'], df['Temperature'], color='tab:red', label='Temperature')
ax1.tick_params(axis='y', labelcolor='tab:red')

# Create a second y-axis for precipitation
ax2 = ax1.twinx()
ax2.set_ylabel('Precipitation (mm)', color='tab:blue')
ax2.bar(df['Date'], df['Precipitation'], color='tab:blue', alpha=0.3, label='Precipitation')
ax2.tick_params(axis='y', labelcolor='tab:blue')

# Title and layout
plt.title('Weather Data: Temperature and Precipitation Over One Year')
plt.tight_layout()
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# Create fake data
np.random.seed(42)
data = {
    'Gender': np.random.choice(['Male', 'Female'], size=200),
    'Score': np.random.normal(loc=75, scale=10, size=200)  # Mean score of 75, std dev of 10
}

df_seaborn = pd.DataFrame(data)

# Create a boxplot to visualize the distribution of scores by gender
plt.figure(figsize=(8, 6))
sns.boxplot(x='Gender', y='Score', data=df_seaborn)
plt.title('Score Distribution by Gender')
plt.xlabel('Gender')
plt.ylabel('Score')
plt.show()
```

```python
import pandas as pd
import numpy as np

# Create fake sales data
np.random.seed(42)
months = pd.date_range(start='2023-01-01', periods=12, freq='M')
sales_data = {
    'Month': months,
    'Total_Sales': np.random.randint(1000, 5000, size=12),  # Random sales between 1000 and 5000
    'Profit': np.random.randint(100, 1000, size=12)  # Random profit between 100 and 1000
}

df_sales = pd.DataFrame(sales_data)
df_sales.set_index('Month', inplace=True)

# Plot Total Sales
plt.figure(figsize=(10, 6))
df_sales['Total_Sales'].plot(kind='line', marker='o', color='blue', title='Monthly Total Sales')
plt.xlabel('Month')
plt.ylabel('Total Sales ($)')
plt.grid()
plt.xticks(rotation=45)  # Rotate x-axis labels for better visibility
plt.show()

# Plot Profit as a bar chart
plt.figure(figsize=(10, 6))
df_sales['Profit'].plot(kind='bar', color='green', title='Monthly Profit')
plt.xlabel('Month')
plt.ylabel('Profit ($)')
plt.xticks(rotation=45)
plt.grid(axis='y')
plt.show()
```
