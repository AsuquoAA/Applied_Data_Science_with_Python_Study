Array creation

```python
#a single dimensional array
a = np.array([1,2,3])
print(a)
print(a.ndim)#number of dimensions of an array using the ndim attribute
```

```python
#a multi-dimensional array
b = np.array([[1,2,3],[4,5,6]])
print(b)
b.shape#gives us the number of rows and number of columns
```

```python
print(a.dtype)#type of items in the arrays
c = np.array([2.2,4,1.1])
c.dtype.name
```

```python
c
```

```python
#creating an array with initial such as zeroes and ones or even random numbers, especially when we know the shape
d = np.zeros((2,3))
print(d)
e = np.ones((2,3))
print(e)
f = np.random.rand(2,3)
print(f)
```

```python
#create a sequence or a range of numbers in an array
g = np.arange(10,50,2)
print(g)
#for a sequence of floats
h = np.linspace(0,7,15)
print(h)
```

Array operations

```python
#arithmetic operations
a = np.array([10,20,30,40])
b = np.array([1,2,3,4])
c = a+b
print(c)
d = a-b
print(d)
e = a*b
print(e)
f = a/b
print(f)
```

```python
farenheit = np.array([32,212,68,98])
celsius = (farenheit - 32) * 5/9
print(celsius)
print(celsius>50)
print(celsius%2==0)
```

```python
#product-wise or matric multiplication, we use @ instead of *
A = np.array([[1,1],[0,1]])
B = np.array([[2,0],[3,4]])
print(A@B)
```

```python
array1 = np.array([[1,2,3],[4,5,6]])
array2 = np.array([[7.1,8.2,9.1],[10.4,11.2,12.3]])
array3 = array1+array2
print(array3)
print(array3.dtype.name)
print(array3.max())
print(array3.min())
print(array3.sum())
print(array3.mean())
```

```python
#creating an array from 1-15 for a 3 by 5 matrix
b = np.arange(1,16,1).reshape(3,5)
print(b)
```
