```python
import pandas as pd
df = pd.read_csv('presidents.csv')
df.head(10)
```

```python
#Clean up President column into first name and last name
#first method uses regex

#first make a copy of the President Column
df['First']=df['President']
#Then we call the replace() and just have a pattern that matches the last name and
#set it to an empty string leaving only the first name
df['First'] = df['First'].replace('[ ]\w+','',regex = True)
df.head(10)
```

```python
#This works but its kinda gross and slow
del(df['First'])
```

```python
def splitname(row):
    row['First'] = row['President'].split(' ')[0]
    row['Last'] = row['President'].split(' ')[-1]
    return row

df = df.apply(splitname, axis='columns')
df.head(10)    
```

```python
# Pretty questionable as to whether that is less gross, but it achieves the result and
#  I find that I use the apply() function regularly in my work. The pandas series has 
# a couple of other nice convenience functions though, and the next I would like to 
# touch on is called .extract(). Lets drop our firstname and lastname.
#del(df['First'])
#del(df['Last'])

#Extract takes a regex and requires you to set capture groups that correspond to the 
#output columns youre interested in
pattern = r'(^[\w]+)(?:.*\s)([\w]+$)'
df['President'].str.extract(pattern).head(10)
```

```python
pattern = r'(?P<First>[\w]+)(?:.*\s)(?P<Last>[\w]+$)'
df['President'].str.extract(pattern).head(10)
```

```python
#now we call extract
names = df['President'].str.extract(pattern)
names.head()
```

```python
#Now let's clean up the born column
df['Born'] = df['Born'].str.extract(r'([\w]{3} [\d]{1,2}, [\d]{4})')
df['Born'].head(10)
```

```python
#update the column to date time
df['Born']=pd.to_datetime(df['Born'])
df['Born'].head(10)
```

```python
import pandas as pd
import numpy as np
import timeit

df=pd.read_csv('census.csv')
df.head(20)
```

```python
#First method: Method chaining, the so-called pandorable way to write the code
(df.where(df['SUMLEV']==50)).dropna().set_index(['STNAME','CTYNAME']).rename(columns={'ESTIMATESBASE2010': 'Estimates Base 2010'})
```

```python
#Second method: traditional code and more readable
df = df[df['SUMLEV']==50]
df.set_index(['STNAME','CTYNAME']).rename(columns={'ESTIMATESBASE2010': 'Estimates Base 2010'})
```

```python
def first_approach():
    global df
    return df.where(df['SUMLEV']==50).dropna().set_index(['STNAME','CTYNAME']).rename(columns={'ESTIMATESBASE2010':'Estimates Base 2010'})

timeit.timeit(first_approach,number=1)
```

```python
df
```

```python
def second_approach():
    global df
    new_df = df[df['SUMLEV']==50]
    new_df.set_index(['STNAME','CTYNAME'],inplace=True)
    return new_df.rename(columns={'ESTIMATESBASE2010': 'Estimates Base 2010'})
timeit.timeit(second_approach,number=1)    
```

```python
#hence, second approach is much faster

def min_max(row):
    data = row[['POPESTIMATE2010',
                'POPESTIMATE2011',
                'POPESTIMATE2012',
                'POPESTIMATE2013',
                'POPESTIMATE2014',
                'POPESTIMATE2015']]
    return pd.Series({'min':np.min(data),'max':np.max(data)})            
```

```python
def min_max(row):
    data = row[['POPESTIMATE2010',
                'POPESTIMATE2011',
                'POPESTIMATE2012',
                'POPESTIMATE2013',
                'POPESTIMATE2014',
                'POPESTIMATE2015']]
    row['max'] = np.max(data)
    row['min'] = np.min(data)
    return row
df.apply(min_max, axis=1)
```

```python
#One line example of how you might calculate the max of the columns using the apply 
# function with lambda. 
rows = ['POPESTIMATE2010',
        'POPESTIMATE2011',
        'POPESTIMATE2012',
        'POPESTIMATE2013',
        'POPESTIMATE2014',
        'POPESTIMATE2015']

df.apply(lambda x: np.max(x[rows]),axis='columns')        
```

```python
def get_state_region(x):
    northeast = ['Connecticut', 'Maine', 'Massachusetts', 'New Hampshire', 
                 'Rhode Island','Vermont','New York','New Jersey','Pennsylvania']
    midwest = ['Illinois','Indiana','Michigan','Ohio','Wisconsin','Iowa',
               'Kansas','Minnesota','Missouri','Nebraska','North Dakota',
               'South Dakota']
    south = ['Delaware','Florida','Georgia','Maryland','North Carolina',
             'South Carolina','Virginia','District of Columbia','West Virginia',
             'Alabama','Kentucky','Mississippi','Tennessee','Arkansas',
             'Louisiana','Oklahoma','Texas']
    west = ['Arizona','Colorado','Idaho','Montana','Nevada','New Mexico','Utah',
            'Wyoming','Alaska','California','Hawaii','Oregon','Washington']
    
    if x in northeast:
        return 'Northeast'
    elif x in midwest:
        return 'Midwest' 
    elif x in south:
        return 'South'
    else:
        return 'West'           
```

```python
#Now we have customized a function, let's say we want to create a new column called
#Region which shows the state's region

df['state region'] = df['STNAME'].apply(lambda x: get_state_region(x))
#or you can use 
df['state region'] = df['STNAME'].apply(get_state_region)
```
